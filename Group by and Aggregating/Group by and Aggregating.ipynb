### Aggregations with Group By

```python
df.groupby('education')['outstanding_balance'].max() #seeing the max or min
```

```python
df.groupby('education')['outstanding_balance'].sum().sort_values(ascending = False) #sorting the values
```

```python
df.groupby('education')['outstanding_balance'].count().sort_values(ascending = False) #sorting the values
```

```python
df.groupby('education')['outstanding_balance'].median().sort_values(ascending = False) #seeing the median
```

```python
df.groupby('education').agg({'outstanding_balance' : ['sum','count']}) #putting two aggregate fucnctions in one DF
```

```python
output = df.groupby('education').agg({'outstanding_balance' : ['sum','count']})
```

```python
#finding the average outstanding_balance per person
output['average'] = output['outstanding_balance']['sum']/output['outstanding_balance']['count']*100
```

```python
output.sort_values(by = ('average'), ascending = False)
```

```python
#making aggregate functions while having only one title per column

df.groupby('education').agg(sum_balance = ('outstanding_balance','sum'),count_balance = ('outstanding_balance','count'))
```
