### Correlation

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
MSRP = pd.read_csv(r'/Users/georgiosskourlis/Documents/Analyst Builder/Pandas for Data Analysis/car_sales.csv')

MSRP
```

```python
MSRP.head()
```

```python
# Select only numeric columns for correlation analysis
numeric_cols = MSRP.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_msrp = numeric_cols.corr()
corr_msrp
```

```python
sns.heatmap(corr_msrp, annot=True, cmap='coolwarm')
```

```python
sns.scatterplot(data = MSRP, x = 'Engine HP', y = 'MSRP')

sns.regplot(data = MSRP, x = 'Engine HP', y = 'MSRP', scatter = False, color = 'red')

plt.show()
```
