### Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'First Label', color = 'green', marker = '^',  markersize = 10, linestyle = '--', markeredgecolor = 'blue' )

plt.title('Matplotlib Viz', fontdict  = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])


plt.legend()

plt.show()
```

```python
x= [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'First Label', color = 'green', marker = '^',  markersize = 15, linestyle = '--', markeredgecolor = 'blue' )


z = [1,5,10]
plt.plot(x,z, label = 'Second Label', color = 'red', marker = '.',  markersize = 15, linestyle = '--', markeredgecolor = 'green' )

plt.title('Matplotlib Viz', fontdict  = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])


plt.legend()

plt.show()
```

```python
x= [1,2,3]
y = [5,10,20]


plt.figure(figsize = (20,5))

plt.plot(x,y, label = 'First Label', color = 'green', marker = '^',  markersize = 15, linestyle = '--', markeredgecolor = 'blue' )


z = [1,5,10]
plt.plot(x,z, label = 'Second Label', color = 'red', marker = '.',  markersize = 15, linestyle = '--', markeredgecolor = 'green' )

plt.title('Matplotlib Viz', fontdict  = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])


plt.legend()

#plt.savefig()

plt.show()

```
