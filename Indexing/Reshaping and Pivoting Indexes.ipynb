```python
df_world = pd.read_csv(r'/Users/georgiosskourlis/Documents/Analyst Builder/Pandas for Data Analysis/WorldContinents.csv',
                    encoding = 'ISO-8859-1')

df_world
```

```python
#Pivoting the table with these 3 conditions : index = '', columns = '',values = ''

df_world.pivot(index = 'country', columns = 'continent',values = 'sub_region')
```

```python
# A different view of the data while pivoting

df_world.pivot(index = 'country', columns = 'continent',values = 'Unnamed: 4')
```
